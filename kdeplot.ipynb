```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# kdeplot -> kernel density estimation
# rather than using discrete bins, a KDE plot smooths the observations wiht a
# Gaussiam kernel , producing a continous density estimate
tips = sns.load_dataset("tips")
# axes level function
sns.kdeplot(data=tips, x="total_bill")
```

```python
# hue -> fill
sns.displot(data=tips, x="total_bill",kind='kde',hue='sex',fill=True)
```
