#### Regression Plots (relationship plots)

These are super important in Machine Learning/EDA because they help us visualize relationships between two continuous variables and optionally fit a regression model (usually linear).

A regression plot = scatter plot + fitted regression line (with confidence interval).

In Seaborn, the key functions are:

regplot() → basic regression plot (axes-level).

lmplot() → higher-level regression plot (figure-level, supports facets).

```python
import seaborn as sns
import numpy as np
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
# Scatter points show actual data, blue line shows fitted linear regression,
# shaded area = confidence interval (95% by default).
```

Important Parameters


(A) Data & Fitting

x, y, data → variables.

order=2 → polynomial regression (quadratic fit).

logx=True → log-transform x-axis before fitting.

x_estimator=np.mean → reduce scatter by plotting aggregated x values.


(B) Scatter Control

scatter=True/False → show/hide scatter points.

# lmplot

Wrapper around regplot but figure-level (like catplot).

Allows grouping/faceting with col, row, hue.

More powerful for multi-plot comparisons.

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
# Separate regression lines for male and female customers.
# Figure level function
# hue parameter is only allowed in lmplot , not in regplot
```

# Parameters (extra compared to regplot)

hue → multiple regression lines for different categories.

col / row → facet plots by category.

markers=["o","s"] → different markers for groups.

palette="Set1" → color theme.

height=4, aspect=1 → figure size control.

# What is a Residual Plot?

A residual = difference between actual data point and predicted value (from regression line).

👉 residual = y_actual − y_predicted

A residual plot shows x-values vs residuals.

If the regression model is a good fit:

Residuals should look like random noise around zero.

No clear trend should appear.

```python
sns.residplot(x="total_bill", y="tip", data=tips)
# Scatter points represent residuals (difference from regression line). The horizontal dashed
# line at 0 = perfect fit.
```
