### Figure level function -> `catplot`

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

# These are mainly used when one (or more) variables are categorical
# and we want to compare distributions, averages, or relationships.
```

```python
# stripplot -> axes level function

# Plots individual points (jittered).
# Best for small datasets.
# It’s a scatter plot for categorical vs numeric data.
sns.stripplot(data=tips,x='day',y='total_bill',jitter=True)

# jitter=True/False → Spreads dots horizontally so they don’t overlap.
```

```python
# using caplot -> figure level function
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# hue="sex" (color by another category).

sns.stripplot(x="day", y="total_bill", hue="sex", data=tips)
```

2. swarmplot in Seaborn

🔹 What it is:

-> Similar to stripplot, but smarter.

-> In stripplot, even with jitter, points may overlap.

-> In swarmplot, Seaborn arranges dots carefully so they don’t overlap like bees in a swarm 🐝.

-> This makes the distribution shape much clearer.

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue = 'sex')
# Best for small-to-medium datasets (for large datasets, too many dots
# may slow it down).
```
