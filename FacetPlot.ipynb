```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# facet plot -> work only with figure level function
# if we want to plot the multiple graph , according to a particular columns
tips = sns.load_dataset('tips')

# clm basis
# sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')

# row basis
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
# if we want one basis of both clm and row
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='smoker',row='sex')
```

```python
# column wrap , if we have multiple graphs and we want only n number of graphs in a row
gap = px.data.gapminder()
sns.relplot(data=gap,x='gdpPercap',y='lifeExp',kind='scatter',col='continent',col_wrap=2)
```
