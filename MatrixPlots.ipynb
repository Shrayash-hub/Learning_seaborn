### 2. Matrix Plot

- Heatmap
- Clustermap

```python

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

gap = px.data.gapminder()

# Heatmap

# plots rectangular data as a color encoded matrix

# When we want to make a heatmap, our data usually needs to be in a matrix form (rows × columns, like a table). But often in datasets, data is in long form (like a list of rows with categories and values).

# The pivot() function helps reshape data into a matrix/table.

# rows as country . clms as years and values as lifeexp
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# sns.heatmap(temp_df)  # -> axes level function
# since we dont have any other figure level functions corresponding to heatmap
# we can resize it similarly we use to resize the figures in matplotlin
plt.figure(figsize=(10,10))
sns.heatmap(temp_df)
```

```python
# annot
# if we want only for european country

temp_df = gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='summer')
# annot=True → Shows numbers inside the cells.

# cmap="YlGnBu" → Yellow-Green-Blue color theme (you can try "coolwarm", "viridis", etc.).
```

### What is clustermap?

-> A clustermap is like a heatmap, but it automatically performs hierarchical clustering on both the rows and columns of your data.

-> It reorders the rows/columns so that similar patterns are grouped together, making hidden structures visible.

```python
iris = px.data.iris()
sns.clustermap(iris.iloc[:,[0,1,2,3]])
# A dendrogram (tree diagram) is added to show how groups were clustered.
```

Parameters to Know

🔹 Data Scaling

standard_scale=0 → Normalize rows (students compared fairly).

standard_scale=1 → Normalize columns (subjects compared fairly).

Default = no scaling.

🔹 Clustering Control

method="average" or "single" or "complete" → Controls how clustering is done.

metric="euclidean" or "correlation" → Defines the distance/similarity metric.

Why Clustermap in ML/DS?

🔍 To find patterns in high-dimensional data (like gene expression, student performance, or sales trends).

📊 To check feature similarity (which features behave similarly).

🧑‍🤝‍🧑 To cluster similar users/rows (unsupervised learning style).

For example:

In ML preprocessing, you may run sns.clustermap(df.corr()) to group correlated features.
