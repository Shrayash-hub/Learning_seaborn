```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Line plot -> if we want to show continuty in our plot
gap = px.data.gapminder()
# if we want to draw graph in terms of india
# first we have to filter data
temp_df = gap[gap['country']== 'India']

sns.lineplot(data=temp_df, x='year', y='lifeExp') # -> axes level function lineplot()
```

```python
# Drawing line plot using relpplot -> figure level function
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# if we want to plot the data of three diffrent countries side by side
# we can use hue parameter
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df

sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country')
# we can observe here that in figure level function , legend is outside the axes,
# that means it have access to the whole figure
```

```python
# If we also want to show the continents of the country means adding a
# another level of information
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent')
# similarily we can use size to add another level of info
```
