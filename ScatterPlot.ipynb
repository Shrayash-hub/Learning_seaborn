```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
# lets plot a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips) # -> axes level function
# if we plot the same plot using matplotlib we have to justify explicitly that
# what is xlable and ylable , but here it is assigned atomatically
```

```python
# plotting scatter plot using relplot -> figure level function
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')
# square shape figure , when we use figure level function

# seaborn prefer to use figure level functions because we can plot multiple
# types of plot using a single function
```

```python
# scatter plot wiht hue prametere -> legend added automatically
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex')
```

```python
# scatter plot with style paramter
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', style='time', hue ='sex')
# so now we represented 4 level of info using a single functions -> that power of seaborn
```

```python
# scatter plot with size parameter -> again a layer of info added
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', size='size', hue ='sex', style='time')
```
