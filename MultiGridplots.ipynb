```python
import seaborn as sns
import numpy as np
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

these are higher-level plotting objects that let you create multiple subplots for different subsets of data.

Seaborn has three main multi-grid objects:

FacetGrid – general-purpose for mapping any plot across facets.

PairGrid – for plotting pairwise relationships between variables.

JointGrid – for creating customized joint plots (scatter + marginal distributions).

# 1. FacetGrid

 What it does??

Creates a grid of subplots based on row and/or column categories.

Then you can "map" any seaborn/matplotlib function to those subplots.

```python
g = sns.FacetGrid(tips, col="sex", row="time")
g.map(sns.scatterplot, "total_bill", "tip")

# Creates a grid:

# Columns = sex (Male/Female)

# Rows = time (Lunch/Dinner)

# Inside each cell: scatterplot of total_bill vs tip.
```

Important Parameters

row, col → facets along row/column.

hue → add color encoding inside each subplot.

# 2. PairGrid

 What it does

Like FacetGrid, but specifically for pairwise relationships between variables.

You can map:

Lower triangle of the grid.

Upper triangle of the grid.

Diagonal (usually univariate plots like hist/kde).

```python
g = sns.PairGrid(iris, hue="species")
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot, fill=True)
g.map_diag(sns.histplot)
g.add_legend()
# This shows pairwise scatterplots, kde plots on upper triangle, histograms
# on diagonal — all separated by species.
```

```python
# same graph using pair grid
g = sns.PairGrid(data=iris)
g.map(sns.scatterplot)
```

Important Parameters

vars=["sepal_length","petal_length"] → choose subset of variables.

# JointGrid vs JointPlot

What it does

Like sns.jointplot, but more customizable.

Creates a central scatter/reg plot and marginal plots along axes.

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip", hue ='sex')
g.plot(sns.scatterplot, sns.histplot) # joint plot me ye vala option nhi hoga
# Central = scatter, top/bottom = histograms.
```
