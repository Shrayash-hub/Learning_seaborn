```python
import seaborn as sns
import numpy as np
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

# Categorical Estimate Plots

These plots donâ€™t just show raw distributions â€” they show statistical estimates (like mean, confidence intervals, or trends) of a numerical variable across categories.

The main ones are:

* barplot

* pointplot

* countplot





# barplot


ðŸ”¹ What it is:

Plots the mean (or another estimator) of a numerical variable for each category.

By default: shows mean + 95% confidence interval (CI).

Great for comparing averages across categories.

```python
# barplot
sns.barplot(x='sex', y='total_bill', data=tips,hue='smoker')
# going to tell about the average value -> bcz estimator is by default mean

# estimator changes like np.median, np.max
sns.barplot(x='sex', y='total_bill', data=tips,hue='smoker',estimator=np.std)
```

# pointplot
ðŸ”¹ What it is:

Similar to barplot, but instead of bars, it shows points with lines connecting them.

Great for showing trends across categories.

```python
sns.pointplot(x='sex', y='total_bill', data=tips,hue='smoker')
# Useful for comparing trends between groups.
```

# countplot
ðŸ”¹ What it is:

Like a barplot, but instead of showing a mean, it shows the count of observations in each category.

Basically: countplot = barplot with estimator=len.

```python
# countplot -> This is similar to a histogram over a categorical,
# rather than quantitative, variable
sns.countplot(data=tips, x='sex',hue='day')
```

barplot â†’ average values + error bars.

pointplot â†’ trends across categories with points + lines.

countplot â†’ counts of categories.

```python
# faceting using catplot
sns.catplot(x='sex', y='total_bill', data=tips,col='smoker',kind='box',row='time')
```
