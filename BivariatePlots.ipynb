```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

tips = sns.load_dataset("tips")

# Bivariate histogram
# a bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations with each rectangle witht the fill color

sns.histplot(data=tips,x='total_bill',y='tip')
# amount of darkness of block shows the density of points in that portion
```

```python
# bivariate kdeplot
# a bivariate KDE plot smoothes the (x,y) observations with a 2D Gaussian
sns.kdeplot(data=tips,x='total_bill',y='tip')
# where the line are  close to each other , that means density at that area is more
```
